## Wine Comparison of Red and White Wines

```python
## Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Loading the datasets

from numpy import concat


red = pd.read_csv('winequality-red.csv', sep = ";")

white = pd.read_csv('winequality-white.csv', sep = ";")

red['type'] = "Red"
white['type'] = "White"

df = pd.concat([red, white])
```

```python
df.head()
```

```python
df.tail()
```

```python
## Comparing Quality of both the wines

sns.countplot(data = df, x = 'quality', hue = 'type' )
plt.show()
```

```python
## Observations

# 1. White Wine dominates quality 5,6 and 7
# 2. Red Wine dominates quality 5 and 6

## Insights

# White wines have slightly higher quality ratings overall.
```

```python
df
```

```python
## Comparing Alcohol Distribution

sns.boxplot(x = 'type', y = 'alcohol', data = df)
```

```python
## Insights

# White wines show slightly higher alchool variability
```

```python
## Comparing Sugar Distribution

sns.boxplot(x = 'type', y = 'residual sugar', data = df)
```

```python
## Insight

# White wines contains slightly higher sugar levels than red wines. 
```

```python
## Comparing Density Distribution

sns.boxplot(x = 'type', y = 'density', data = df)
```

```python
# Insight

# White wines show higher density on average, likely due to higher sugar levels.
```

```python
# Key Comparison Insights

# White wines tend to have higher sugar levels.
# Red wines tend to have higher acidity levels.
# Alcohol positively affects quality in both wine types.
# White wines appear to have slightly higher quality ratings.
```
